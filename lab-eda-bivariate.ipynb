Part 1: Analyzing Best-Seller Trends Across Product Categories

Objective: Understand the relationship between product categories and their best-seller status.

Crosstab Analysis:

- Create a crosstab between the product category and the isBestSeller status.

- Are there categories where being a best-seller is more prevalent?

Hint: one option is to calculate the proportion of best-sellers for each category and then sort the categories based on this proportion in descending order.

```python
# crosstab between the product category and the isBestSeller status.
crosstab_amazon = pd.crosstab(index=df['category'],
            columns='count')
crosstab_amazon
```

```python
crosstab_result = pd.crosstab(index= df["category"],columns = df["isBestSeller"])
crosstab_result
```

```python
# We can see that categories such as Women or Women's Sports & Outdoors Shoes where being a best-seller is more prevalent
```

Statistical Tests:

- Conduct a Chi-square test to determine if the best-seller distribution is independent of the product category.
- Compute Cram√©r's V to understand the strength of association between best-seller status and category.

```python
# Chi-square test to determine if the best-seller distribution is independent of the product category.
chi2_stats, chi2_pvalue, _, _ = chi2_contingency(crosstab_result)
chi2_pvalue
```

```python
# P-value < 0.05 suggests that there is a significant association between the categorical variables.
# In this case, we have a p-value of 0, so we reject that they are independent - we can say that they are related.
```

```python
# Computing the association between variables in 'crosstab_result' using the "cramer" method
association(crosstab_result, method='cramer')
```

```python
# El 0,122 indica una relaci√≥n pobre entre estas dos variables
```

Visualizations:

Visualize the relationship between product categories and the best-seller status using a stacked bar chart.

```python
# Plotting a stacked bar chart for the 'crosstab_result' data
crosstab_result.plot(kind='bar', stacked=True)
plt.show()
```

Part 2: Exploring Product Prices and Ratings Across Categories and Brands

Objective: Investigate how different product categories influence product prices.

Preliminary Step: Remove outliers in product prices.

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
df_sinoutlier = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
df_sinoutlier
```

Violin Plots:

Use a violin plot to visualize the distribution of price across different product categories. 
- Filter out the top 20 categories based on count for better visualization.
- Which product category tends to have the highest median price? Don't filter here by top categories.

```python
top_20_categories = df_sinoutlier['category'].value_counts().head(20).index
top_20_categories
```

```python
# Filtrar el DataFrame para incluir solo estas 20 categor√≠as
df_top_20 = df_sinoutlier[df_sinoutlier['category'].isin(top_20_categories)]
```

```python
# 4. Crear el Violin Plot
plt.figure(figsize=(12, 8))
sns.violinplot(data=df_top_20, x='category', y='price', palette="coolwarm")
plt.xticks(rotation=90)  # Girar etiquetas de las categor√≠as para mejor legibilidad
plt.title('Distribuci√≥n de Precio en las Principales 20 Categor√≠as de Producto')
plt.xlabel('Categor√≠a')
plt.ylabel('Precio')
plt.show()
```

Bar Charts:

- Create a bar chart comparing the average price of products for the top 10 product categories (based on count).
- Which product category commands the highest average price? Don't filter here by top categories.

```python
df_top10 = df_sinoutlier["category"].value_counts().head(10).index
df_final10 = df_sinoutlier[df_sinoutlier['category'].isin(df_top10)]
sns.barplot(data=df_final10, x="category", y="price", palette="coolwarm")
plt.xticks(rotation=45)
plt.show()
```

```python
# Sports and Outdoors category tends to have the highest average price
```

Box Plots:

- Visualize the distribution of product ratings based on their category using side-by-side box plots. Filter out the top 10 categories based on count for better visualization.
- Which category tends to receive the highest median rating from customers? Don't filter here by top categories.

```python
df_top10_ratings = df_sinoutlier["category"].value_counts().head(10).index
df_final10_ratings= df_sinoutlier[df_sinoutlier['category'].isin(df_top10_ratings)]
sns.boxplot(data=df_final10_ratings, x="category", y="stars", palette="coolwarm")
plt.xticks(rotation=45)
plt.show()
```

Which category tends to receive the highest median rating from customers? Don't filter here by top categories.

Part 3: Investigating the Interplay Between Product Prices and Ratings

Objective: Analyze how product ratings (stars) correlate with product prices.

Correlation Coefficients:

- Calculate the correlation coefficient between price and stars.
- Is there a significant correlation between product price and its rating?

```python
df_sinoutlier.head(3)
```

```python
# Ratings is a continous variable
# Product prices is a continuous variable
# PEARSON
df_sinoutlier['price'].corr(df_sinoutlier['stars']) 
```

```python
# There's a weak correlation between product prices and ratings in Amazon products.
# We check this hypothesis with Spearman correlation coefficient 
df_sinoutlier['price'].corr(df_sinoutlier['stars'], method='spearman')
```
